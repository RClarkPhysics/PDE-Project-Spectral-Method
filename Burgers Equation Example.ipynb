```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from PDE_Burgers import *
```

```python
N = 20
v = 0.01
L = 100
dt = 0.001
TotalT = 500

x = np.linspace(0,2*np.pi,L)
t = np.arange(0,TotalT,dt)

PDE = PDE_Burgers(N,v)
Ini = PDE.GenerateInitialCondition()
Solver = PDE.generate_Solver(Ini,t,L)
```

```python
step = 0

plt.figure(figsize=(20,10))
plt.plot(x,PDE.sample_uN(Solver[step],L))
plt.ylabel('u_N value',fontsize=20)
plt.xlabel('x (ranging from 0 to 2pi)',fontsize=20)
plt.title('Bergers Solution at t= '+str(dt*step),fontsize = 20)
plt.show
```

```python
step = 50000

plt.figure(figsize=(20,10))
plt.plot(x,PDE.sample_uN(Solver[step],L))
plt.ylabel('u_N value',fontsize=20)
plt.xlabel('x (ranging from 0 to 2pi)',fontsize=20)
plt.title('Bergers Solution at t= '+str(dt*step),fontsize = 20)
plt.show
```

```python
Steps = 500
def animatep(i):
    tt = Steps*i
    plt.cla()
    plt.xlim([0, 2*np.pi])
    plt.ylim([0, 1])
    plt.plot(x,PDE.sample_uN(Solver[tt],L))
    plt.ylabel('u_N value',fontsize=20)
    plt.xlabel('x (ranging from 0 to 2pi)',fontsize=20)
    plt.title('Bergers Solution',fontsize = 20)
```

```python
fig = plt.figure(figsize=(12, 6))
ani = FuncAnimation(fig, animatep, frames=1000, interval=1, repeat=False)

ani.save('Burgers_Evolution_v01.mp4',writer = 'ffmpeg',fps=30)
```
